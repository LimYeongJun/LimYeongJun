# Lorenz Equation with scipy.integrate.**solve_ivp**

```
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
from scipy.integrate import solve_ivp
import sympy

def f(t, xyz):
    x, y, z = xyz
    return [sigma * (y - x),
            x * (rho - z) - y,
            x * y - beta * z]               # Lorenz Equation의 형태


xyz0 = [1.0, 2.0, 3.0]
t = (0, 25)

rho, sigma, beta = 28, 8, 8/3.0           # Lorenz Equation의 매개변수 값 설정

xyz1 = solve_ivp(f, t, xyz0, t_eval=np.linspace(0,25,10000))     # solve_ivp 에 매개변수 값 설정

rho, sigma, beta = rho, sigma, 0.5*beta

xyz2 = solve_ivp(f, t, xyz0, t_eval=np.linspace(0,25,10000))

rho, sigma, beta = rho, 2*sigma, 0.5*beta

xyz3 = solve_ivp(f, t, xyz0, t_eval=np.linspace(0,25,10000))

from mpl_toolkits.mplot3d.axes3d import Axes3D                   # 그래프그리기

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(12, 3.5), subplot_kw={'projection': '3d'})

for ax, xyz, c in [(ax1, xyz1, 'r'), (ax2, xyz2, 'b'), (ax3, xyz3, 'g')]:
    ax.plot(xyz.y[0], xyz.y[1], xyz.y[2], c, alpha=0.5)
    ax.set_xlabel('x', fontsize=16)
    ax.set_ylabel('y', fontsize=16)
    ax.set_zlabel('z', fontsize=16)
    ax.set_xticks([-15, 0, 15])
    ax.set_yticks([-20, 0, 20])
    ax.set_zticks([0, 20, 40])
```

# 이와 같이 Lorenz 방정식을 초기값을 이용하여 해를 그래프로 나타낼 수 있다.
